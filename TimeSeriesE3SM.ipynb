### Time Series for Athens, GA and Seattle, WA

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cftime
import pandas as pd
from datetime import datetime
from matplotlib.animation import FuncAnimation
```

```python
#Load .nc file for E3SM.HR and E3SM.MMF
#HR
path = '/Users/emmabenedict/Desktop/CURO 4970R/'
esm_hr = xr.open_dataset(path + 'E3SM.HR.ne120pg2.0001-0010.US.2x2.5.hourly.PRECC.PRECL.nc')
#MMF
path = '/Users/emmabenedict/Desktop/CURO 4970R/'
esm_mmf = xr.open_dataset(path + 'E3SM.MMF.ne30pg2_CRM64x2000.0001-0010.US.2x2.5.hourly.PRECC.PRECL.nc')
```

```python
#esm_hr
#esm_mmf
```

```python
#Load variables
#HR
lat_hr = esm_hr.lat
lon_hr = esm_hr.lon
time_hr = esm_hr.time
precc_hr = esm_hr.PRECC
precl_hr = esm_hr.PRECL
precip_hr = precc_hr + precl_hr
area_hr = esm_hr.area
#MMF
lat_mmf = esm_mmf.lat
lon_mmf = esm_mmf.lon
time_mmf = esm_mmf.time
precc_mmf = esm_mmf.PRECC
precl_mmf = esm_mmf.PRECL
precip_mmf = precc_mmf + precl_mmf
area_mmf = esm_mmf.area
```

```python
#Pick out two locations: Athens, GA & Seattle, WA
hr_athens = precip_hr.sel(lon=276.64,lat=33.95,method="nearest")*1000*60*60
hr_seattle = precip_hr.sel(lon=237.67,lat=47.60,method="nearest")*1000*60*60
mmf_athens = precip_mmf.sel(lon=276.64,lat=33.95,method="nearest")*1000*60*60
mmf_seattle = precip_mmf.sel(lon=237.67,lat=47.60,method="nearest")*1000*60*60
```

```python
avg = np.mean(hr_athens)
avg1 = np.mean(hr_seattle)
avg2 = np.mean(mmf_athens)
avg3 = np.mean(mmf_seattle)
print(avg)
print(avg1)
print(avg2)
print(avg3)
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(hr_athens)
plt.xlabel("Time")
plt.ylabel("Rainfall (mm/hr)")
plt.title("Rain in Athens -- HR")

plt.subplot(2,2,2)
plt.plot(mmf_athens)
plt.xlabel("Time")
plt.ylabel("Rainfall (mm/hr)")
plt.title("Rain in Athens -- MMF")

plt.subplot(2,2,3)
plt.plot(hr_seattle)
plt.xlabel("Time")
plt.ylabel("Rainfall (mm/hr)")
plt.title("Rain in Seattle -- HR")

plt.subplot(2,2,4)
plt.plot(mmf_seattle)
plt.xlabel("Time")
plt.ylabel("Rainfall (mm/hr)")
plt.title("Rain in Seattle -- MMF")

plt.tight_layout()
```
