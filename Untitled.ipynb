### Calculating Percentiles for E3SM Data

```python
# Load Packages
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cftime
import pandas as pd
```

```python
# Load E3SM -- High Res
path = '/Users/emmabenedict/Desktop/CURO 4970R/'
esm_hr_nc = xr.open_dataset(path + 'E3SM.HR.ne120pg2.0001-0010.US.2x2.5.hourly.PRECC.PRECL.nc')
```

```python
# Load Variables
lat = esm_hr_nc.lat
lon = esm_hr_nc.lon
time = esm_hr_nc.time
precc = esm_hr_nc.PRECC
precl = esm_hr_nc.PRECL
precip = precc + precl
area = esm_hr_nc.area
#Add PRECC and PRECL
#CC -- Convective Paramaterization processes
#CL -- Large scale processes
```

```python
#esm_hr_nc
```

```python
# Pick out location to plot precipitation data -- Athens, GA
precip_athens = precip.sel(lon=276.64,lat=33.95,method="nearest")*1000*60*60*24
```

```python
precip_25 = np.percentile(precip, 25)
```

```python
# Put in axis = 0 
athens_25 = np.percentile(precip, 25, axis = 0)

athens_50 = np.percentile(precip_athens, 50)
athens_75 = np.percentile(precip_athens, 75)
athens_99 = np.percentile(precip_athens, 99)
```

```python
athens_25.size
```

```python
athens_25_plot = precip_athens[precip_athens <= athens_25]
athens_50_plot = precip_athens[(precip_athens > athens_25) & (precip_athens <= athens_50)]
athens_75_plot = precip_athens[(precip_athens > athens_50) & (precip_athens <= athens_75)]
athens_99_plot = precip_athens[(precip_athens > athens_75) & (precip_athens <= athens_99)]
```

```python
plt.plot(athens_25_plot)
plt.show()
```
